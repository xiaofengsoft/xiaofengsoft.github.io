绘制置信区间的误差棒图

```python
import plotly.graph_objects as go
import numpy as np
import scipy.stats as stats
```

$$
P(\theta_1 <= \theta <= \theta_2) = 1 - \alpha
$$

```python
# 样本总数
X = stats.norm.rvs(loc=0, scale=1, size=100)
# 均值
mean = np.mean(X)
# 标准差
std = np.std(X, ddof=1)
# 标准误差
standard_error = std / np.sqrt(len(X))
# 95%置信区间
confidence_level = 0.95
z_score = stats.norm.ppf((1 + confidence_level) / 2)
z_score
```

```python
margin_of_error = z_score * standard_error
margin_of_error
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=np.arange(len(X)), y=X, error_y={
    'type': 'data',
    'array': margin_of_error,
    'visible': True
}, mode='markers', name='Error Bar'))
fig.show()
```
