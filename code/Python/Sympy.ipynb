### 打印Latex格式

```python
from sympy import *
init_printing()
y = symbols('y', cls=Function)
x = symbols('x')
eq = Eq(y(x).diff(x, 2), 0)
sol = dsolve(eq, y(x))
display(sol)
```

### 解常微分方程组

解如下的常微分方程组：
$$
\begin{cases}
\frac{dx_1}{dt} = 2x_1 - 3x_2 + 3x_3 , x_1(0) = 1\\ 
\frac{dx_2}{dt} = 4x_1 - 5x_2 + 3x_3 , x_2(0) = 2\\
\frac{dx_3}{dt} = 4x_1 - 4x_2 + 2x_3 , x_3(0) = 3
\end{cases}
$$

```python
t = symbols('t')
x1, x2, x3 = symbols('x1 x2 x3', cls=Function)
eq = [
    x1(t).diff(t, 1) - 2*x1(t) + 3*x2(t) - 3*x3(t),
    x2(t).diff(t, 1) - 4*x1(t) + 5*x2(t) - 3*x3(t),
    x3(t).diff(t, 1) - 4*x1(t) + 4*x2(t) - 2*x3(t)
]
con = {
    x1(0): 1,
    x2(0): 0,
    x3(0): 0
}
sol = dsolve(eq, [x1(t), x2(t), x3(t)], ics=con)
display(sol[0])
display(sol[1])
display(sol[2])
```

矩阵解法

```python
x = Matrix([x1(t), x2(t), x3(t)])
A = Matrix([
    [2, -3, 3],
    [4, -5, 3],
    [4, -4, 2]
])
eq = x.diff(t) - A*x
sol = dsolve(eq, ics={x1(0): 1, x2(0): 2, x3(0): 3})
display(sol)
```
